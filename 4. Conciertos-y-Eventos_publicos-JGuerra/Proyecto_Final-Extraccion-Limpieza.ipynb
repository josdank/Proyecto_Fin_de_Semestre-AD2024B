```python
# Carga de los CSVs
files = ['actions.csv', 'events_World_Cup.csv', 'features.csv', 'games.csv', 'player_games.csv', 'playerank.csv', 'players.csv']
dfs = []

# Cargar y mostrar los primeros 10 registros de cada archivo
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)
    print(f"Primeros 10 registros de {file}:")
    print(df.head(10)) 
    print("\n")

# Verifica que el total de registros combinados sea de al menos 200,000
total_rows = sum([df.shape[0] for df in dfs])
print(f"\nTotal de registros en todos los CSVs: {total_rows}")
```

```python
files = ['actions.csv', 'events_World_Cup.csv', 'features.csv', 'games.csv', 'player_games.csv', 'playerank.csv', 'players.csv']
dfs = []

for file in files:
    df = pd.read_csv(file)
    dfs.append(df)
    print(f"Archivo cargado: {file}")
```

```python
import os
import pandas as pd
# Carpeta donde guardar los archivos limpios
output_folder = 'archivos_limpios'

# Verificar si la carpeta existe, si no, crearla
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Limpieza: Eliminar duplicados y columnas vacÃ­as
limpieza_dfs = []

for i, df in enumerate(dfs):
    # Eliminar duplicados
    df_limpio = df.drop_duplicates()
    
    # Eliminar columnas vacÃ­as
    df_limpio = df_limpio.dropna(axis=1, how='all')
    
    # Guardar el dataframe limpio en una lista
    limpieza_dfs.append(df_limpio)
    
    # Guardar el archivo CSV limpio en la carpeta especificada
    cleaned_file = os.path.join(output_folder, f'limpio_{files[i]}')
    df_limpio.to_csv(cleaned_file, index=False)
    
    print(f"Archivo limpio guardado en: {cleaned_file}")
```
