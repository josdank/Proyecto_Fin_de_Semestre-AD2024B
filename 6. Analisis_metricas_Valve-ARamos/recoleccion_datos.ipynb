```python
def steam_reviews(csv_path, app_id):
    try:
        data = pd.read_csv(csv_path)
        reviews = data[data["app_id"] == app_id]
        return reviews 
    
    except FileNotFoundError:
        print("El archivo no fue encontrado")
        return None
    except KeyError:
        print("El archivo no tiene columna 'app_id'")
        return None
```

```python
app_id = 70 #half life 1

data = steam_reviews(csv_path, app_id)

if data is not None:
    client = MongoClient("mongodb://localhost:27017/")
    db = client["steam_reviews"]
    collection = db["hl_reviews"]

    result = collection.insert_many(data.to_dict("records"))
    print(f"Datos insertados en MongoDB con éxito. Total de reviews para el juego con id {app_id}: {len(result.inserted_ids)}")

    print(data.head())
else:
    print("No se pudieron cargar los datos del CSV")
```

```python
cs_dataset_path = "csgo_steam_reviews.csv"

def cs_reviews(cs_dataset_path):
    try:
        data = pd.read_csv(cs_dataset_path)

        cs2_data = pd.DataFrame({
            "app_id" : 730,
            "app_name" : "Counter-Strike 2",
            "review_text": data["review"],
            "review_score": data["voted_up"].apply(lambda x: 1 if x else 0),
            "review_votes": data["votes_up"]
        })
        
        cs2_data = cs2_data.dropna()
        
        return cs2_data

    except FileNotFoundError:
        print("El archivo no fue encontrado")
        return None
    except KeyError as e:
        print(f"Error: falta una columna en el CSV - {e}")
        return None
```

```python
data = cs_reviews(cs_dataset_path)

if data is not None:
    client = MongoClient("mongodb://localhost:27017/")
    db = client["steam_reviews"]
    collection = db["cs2_reviews"]

    result = collection.insert_many(data.to_dict("records"))
    print(f"Datos insertados en MongoDB con éxito. Total de registros insertados: {len(result.inserted_ids)}")

    data.head()
    
else:
    print("No se pudieron cargar los datos del CSV")
```

```python
def unir_colecciones():
    try:
        client = MongoClient("mongodb://localhost:27017/")
        db = client["steam_reviews"]
        collections = db.list_collection_names()
        
        excepto = ["cs2demo_reviews", "all_reviews"]
        dataFrames = []
        
        for collection  in collections:
            if collection in excepto:
                print(f"Coleccion no añadida: {collection}")
                continue
            
            print(f"Procesando coleccion: {collection}")
            datos= list(db[collection].find())
            if datos:
                df = pd.DataFrame(datos)
                df["source_collection"] = collection
                dataFrames.append(df)
        
        if dataFrames:
            union_df = pd.concat(dataFrames, ignore_index=True)
            print("Todas las colecciones fueron unidas exitosamente")
            
            data_dict = union_df.to_dict(orient="records")
            db["reseñas_totales"].insert_many(data_dict)
            print(f"Datos guarados en la nueva coleccion")
        
        else: 
            print("No se encontraron datos en las colecciones")
            
    except Exception as e:
        print(f"Ocurrio un error en: {e}")
```

```python
unir_colecciones()
```
