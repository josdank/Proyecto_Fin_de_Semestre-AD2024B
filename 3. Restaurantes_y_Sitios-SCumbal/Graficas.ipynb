```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Distribución de calificaciones
df = pd.read_csv("Restaurantes_base.csv")
plt.figure(figsize=(8, 5))
df['stars'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Distribución de Calificaciones")
plt.xlabel("Estrellas")
plt.ylabel("Cantidad de Negocios")
plt.show()
```

```python
# 2. Top ciudades por cantidad de negocios
plt.figure(figsize=(10, 5))
df['city'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Ciudades con Más Negocios")
plt.xlabel("Ciudad")
plt.ylabel("Cantidad de Negocios")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Relación entre calificaciones y número de reseñas
plt.figure(figsize=(8, 5))
plt.scatter(df['stars'], df['review_count'], alpha=0.5, color='purple')
plt.title("Relación entre Calificaciones y Número de Reseñas")
plt.xlabel("Estrellas")
plt.ylabel("Cantidad de Reseñas")
plt.show()
```

```python
# 4. Proporción de negocios abiertos y cerrados
plt.figure(figsize=(6, 6))
df['is_open'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'salmon'], labels=['Abiertos', 'Cerrados'])
plt.title("Proporción de Negocios Abiertos vs Cerrados")
plt.ylabel("")
plt.show()
```
